#DATA TRANSFORMATION

```python
df_arg.head(5)
```

```python
#create new collumn
df_arg["occ_pct"]=df_arg["successful_bookings"]/df_arg["capacity"]
df_arg.head(5)
```

```python
#to make oct_pct look beter we do the round off
df_arg["occ_pct"]=df_arg["occ_pct"].apply(lambda x: round(x*100,1))
df_arg.head(6)
```

#INSITE GENTRATION

# WHAT IS AVRAGE OCCUPANCY RATE IN EACH OF THE ROOM CATAGORY 

```python
df_arg.groupby("room_category")["occ_pct"].mean().round(2)
```

```python
df=pd.merge(df_arg ,df_room,left_on="room_category",right_on="room_id")
df.tail(5)
```

```python
df.groupby("room_class")["occ_pct"].mean().round(2)
```

```python
df.drop("capacity",axis=1,inplace=True)
df.head(5)
```
